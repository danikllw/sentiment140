```python
import os
import re
import time
import random
import joblib
import mlflow
import numpy as np
import pandas as pd

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

SEED = 42
MEMBER_NAME = "Daniel Wilches"
MLFLOW_URI = "http://44.203.179.96:5000"
EXPERIMENT_NAME = "sentiment140_lab_v2"
REPO_ROOT = "sentiment140"

random.seed(SEED)
np.random.seed(SEED)

mlflow.set_tracking_uri(MLFLOW_URI)
mlflow.set_experiment(EXPERIMENT_NAME)

def basic_clean(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

print("Cargando dataset desde Hugging Face Datasets...")
dataset = load_dataset("adilbekovich/Sentiment140Twitter")

train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

train_df = train_df.copy()
test_df = test_df.copy()

train_df["text_clean"] = train_df["text"].apply(basic_clean)
test_df["text_clean"] = test_df["text"].apply(basic_clean)

# test se reserva; solo partimos train para validaciГіn
X_train, X_valid, y_train, y_valid = train_test_split(
    train_df["text_clean"],
    train_df["label"],
    test_size=0.2,
    random_state=SEED,
    stratify=train_df["label"]
)

baseline = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("clf", MultinomialNB())
])

print("Entrenando baseline...")
start = time.time()
baseline.fit(X_train, y_train)
elapsed = time.time() - start

preds = baseline.predict(X_valid)
acc = accuracy_score(y_valid, preds)

model_path = f"{REPO_ROOT}/models/best_model.pkl"
os.makedirs(os.path.dirname(model_path), exist_ok=True)
joblib.dump(baseline, model_path)

with mlflow.start_run(run_name="baseline_bow_nb_sagemaker"):
    mlflow.set_tag("member", MEMBER_NAME)
    mlflow.set_tag("stage", "baseline")
    mlflow.set_tag("dataset", "adilbekovich/Sentiment140Twitter")
    mlflow.set_tag("environment", "sagemaker")

    mlflow.log_param("seed", SEED)
    mlflow.log_param("preprocessing", "basic_clean")
    mlflow.log_param("encoding", "bow")
    mlflow.log_param("model", "MultinomialNB")
    mlflow.log_param("data_source", "HuggingFace: adilbekovich/Sentiment140Twitter")

    mlflow.log_metric("val_accuracy", float(acc))
    mlflow.log_metric("train_time_sec", float(elapsed))

    mlflow.log_artifact(model_path, artifact_path="model")

print("Baseline accuracy:", acc)
print("Artifact guardado en:", model_path)
print("Run registrado en MLflow con tag member =", MEMBER_NAME)
```
